The linear model corresponds to

Where

- $Y$: Harvest in kilograms

By running the above code the least squares estimators for the parameters are calculated.

- $\hat\beta_0$: 97.77938
- $\hat\beta_1$: 0.07440

Therefore, it's possible to say that the explanatory variable has nothing to do with the response variable and so the `p-value` is less than 0.05 so we can see the estimated parameters are good.
