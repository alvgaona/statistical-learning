Using [graduates.csv](../data/graduates.csv) we have calculated its mean, median, truncated mean (10%) as well as standard deviation, interquartile range and MAD.

```r
## Read graduates grades subscribed in posgraduate program at the Department of Industrial
## Engineering in University of Berkeley.
graduates <- read.csv("../data/graduates.csv")
head(graduates, n = 5)
```

```r
df <- data.frame(mean = mean(grades),
                 median = median(grades),
                 truncated_mean = mean(grades, trim = 0.1))
head(df)
```

```r
df <- data.frame(sd = sd(grades),
                 iqr = IQR(grades),
                 mad = mad(grades))
head(df)
```

In the boxplot we can see that there are no outliers and the previous parameters are involved in it.
The median is the value in the middle of the box and in this case matches the mean.
The IQR is the size of the box and it's related to the standard deviation by a factor of 0.6745.

```r
boxplot(grades, main = "Grades boxplot", xlab = "Graduates", ylab = "Grades")
```

According to the normal Q-Q plot we can assume it corresponds to a gaussian distribution.

```r
qqnorm(grades, main = "Grades | Normal Q-Q Plot"); qqline(grades) 
```
