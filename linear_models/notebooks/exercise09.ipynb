## Standard Normal Distribution

$$f(x)=\frac{1}{\sqrt{2\pi\sigma}}e^{-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^2} \{-\infty < x < \infty\}$$

With

- $\mu = 0$
- $\sigma = 1$

Generated 1000 samples

```r
x <- rnorm(1000)

hist(x, main = "Normal histogram")
boxplot(x, main = "Normal boxplot",
        ylab = "Sample quantities",
        xlab = "Groups",
        col = rgb(0.1,0.1,0.9,0.1))
qqnorm(x, main = "Normal Q-Q Plot | Normal dist."); qqline(x, col = 2)
```

## Binomial distribution

The binomial distribution is formally defined as

$$p(k) = \binom{n}{k}p^nq^{n-k}$$

- $n$: number of independent $X_i \sim Bernoulli(p)$ trials
- $k$: number of successes
- $p$: probability of k successes

In this particular scenario I have chosen

- $n = 1000$
- $k = 10$
- $p = 0.4$

```r
x <- rbinom(1000,10,0.4)

hist(x, main = "Binomial histogram")
boxplot(x, main = "Binomial boxplot",
        ylab = "Sample qunatities",
        xlab = "Groups",
        col = rgb(0.7,0.3,0.4,0.9))
qqnorm(x, main = "Normal Q-Q Plot | Binomial dist."); qqline(x, col = 2)
```

## Chi-squared distribution

The Chi-squared distribution is formally defined as

$$f(x) = \frac{1}{2^{k/2}\Gamma\left(\frac{k}{2}\right)}x^{\frac{k}{2}-1}e^{-\frac{x}{2}} \left\{ x \geq \infty\right\}$$

Where

$k$: number of degrees of freedom

In this particular scenario I have chosen

- $N$: 1000 samples
- $k$: 50 degrees of freedom

```r
## Histogram, boxplot and Q-Qnorm for a chi-square distribution
x <- rchisq(1000,50)

hist(x, main = "Chi-square histogram")
boxplot(x, main = "Chi-square boxplot",
        ylab = "Sample qunatities",
        xlab = "Groups",
        col = rgb(0.5,0.2,0.4,0.6))
qqnorm(x, main = "Normal Q-Q Plot | Chisq dist."); qqline(x, col = 2)
```

## Fischer-Snedecor distribution

The F distribution is formally defined as

$$f(x) = \frac{\sqrt{\frac{{d_1x}^{d_1}d_2^{d_2}}{(d_1x+d_2)^{d_1+d_2}}}}{x B\left(\frac{d_1}{2},\frac{d_2}{2}\right)} \left\{x \geq \infty\right\}$$

Where

- $d_1$: number of freedoms
- $d_2$: number of freedoms
- $B$: Beta function

In this scenario I have chosen

- $N$: 1000 samples
- $d_1$: 90
- $d_2$: 40

```r
x <- rf(1000,90,40)

hist(x, main = "F histogram")
boxplot(x, main = "F boxplot",
        ylab = "Sample qunatities",
        xlab = "Groups",
        col = rgb(0.8,0.1,0.3,0.6))
qqnorm(x, main = "Normal Q-Q Plot | F dist."); qqline(x, col = 2)
```

## Gamma distribution

The Gamma distribution is formally defined as

$$f(x) = \frac{1}{\Gamma(k)\theta^k}x^{k-1}e^{\frac{x}{\theta}} \left\{ x \geq 0 \right\}$$

Where

- $k$: shape
- $\theta$: scale

In this scenario I have chosen

- $N$: 1000 samples
- $k$: 0.7
- $\theta$: 1

```r
x <- rgamma(1000,0.7)

hist(x, main = "Gamma histogram")
boxplot(x, main = "Gamma boxplot",
        ylab = "Sample qunatities",
        xlab = "Groups",
        col = rgb(0.1,0.1,0.7,0.5))
qqnorm(x, main = "Normal Q-Q Plot | Gamma dist."); qqline(x, col = 2)
```
